# About the Data 
Input variables (based on physicochemical tests):
* 1 - fixed acidity
* 2 - volatile acidity
* 3 - citric acid
* 4 - residual sugar
* 5 - chlorides
* 6 - free sulfur dioxide
* 7 - total sulfur dioxide
* 8 - density
* 9 - pH
* 10 - sulphates
* 11 - alcohol
* Output variable (based on sensory data):
* 12 - quality (score between 0 and 10)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("winequality-red.csv")
df
```

```python
data=df.drop('quality',1)
```

```python
data
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_scaled=sc.fit_transform(data)
```

```python
data_sc=pd.DataFrame(data_scaled,index=data.index,columns=data.columns)
data_sc
```

```python
for i in data.columns:
    sns.boxplot(data[i])
    plt.show()
```

```python
data_noOutliers=data.copy()
data_noOutliers
```

```python
for i in data_noOutliers.columns:
    q1=data_noOutliers[i].quantile(0.25)
    q3=data_noOutliers[i].quantile(0.75)
    iqr=q3-q1
    ub=q3 + 1.5*iqr 
    lb=q1 - 1.5*iqr
    uc=data_noOutliers[i].quantile(0.99) #upper cap
    lc=data_noOutliers[i].quantile(0.01) #lower cap
    for ind1 in data_noOutliers[i].index:
        if data_noOutliers.loc[ind1, i] >ub:            
            data_noOutliers.loc[ind1, i] =uc
        if data_noOutliers.loc[ind1, i] < lb:
            data_noOutliers.loc[ind1, i] =lc
            
```

```python
for i in data_noOutliers.columns:
    sns.boxplot(data_noOutliers[i])
    plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
clust_mod=KMeans(n_clusters=3,random_state=0)
clust_mod.fit(data_sc)
clust_mod
```

```python
wcsse=[]
cl=[1,2,3,4,5,6,7,8]
for i in cl:
    mod=KMeans(n_clusters=i,random_state=0)
    mod.fit(data_sc)
    print('K:',i)
    print(mod.inertia_) #inertia value
    wcsse.append(mod.inertia_)
```

```python
plt.plot(cl,wcsse)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data_sc,clust_mod.labels_)

#Which k is having highest silhoutte score
```

```python
sil1=[]
cl=[2,3,4,5,6,7,8] # 1 should not be there as we cant calculate the silhoutte score with one cluster, we need min of 2
for k in cl:       # with one cluster we can only find wcsse
    mod=KMeans(n_clusters=k)
    mod.fit(data_sc)
    sil=silhouette_score(data_sc,mod.labels_)
    sil1.append(sil)
sil1
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

model=KMeans(n_clusters=2,random_state=1) #from elbow curve the best k value we got as k=4
visualizer=SilhouetteVisualizer(model,colors='yellowbrick')
visualizer.fit(data_sc) #dats with outlier
visualizer.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
out=data_sc['Labels']
inp=data_sc.drop('Labels',1)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(inp,out,test_size=0.3,random_state=1)
```

```python
lr_mod=lr.fit(xtrain,ytrain)
lr_mod
```

```python
ypred=lr_mod.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```
