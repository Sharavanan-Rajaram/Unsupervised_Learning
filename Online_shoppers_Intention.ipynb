```python
data=pd.get_dummies(data,columns=['Month','VisitorType'])
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
data_sc1=sc.fit_transform(data)
data_sc1=pd.DataFrame(data_sc1,index=data.index,columns=data.columns)
```

```python
for i in data_sc1.columns:
    q1=data_sc1[i].quantile(0.25)
    q3=data_sc1[i].quantile(0.75)
    iqr=q3-q1
    ub=q3 + 1.5*iqr  #3 or 4.5 and we can play with this 
    lb=q1 - 1.5*iqr  #3 or 4.5 and we can play with this 
    uc=data_sc1[i].quantile(0.99) #upper cap
    lc=data_sc1[i].quantile(0.01) #lower cap
    for ind1 in data_sc1[i].index:
        if data_sc1.loc[ind1, i] >ub:            
            data_sc1.loc[ind1, i] =uc
        if data_sc1.loc[ind1, i] < lb:
            data_sc1.loc[ind1, i] =lc
```

```python
#Model Building
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1,8):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_sc1)
    wcss.append(kmeans.inertia_)
    
# Visualization of k values:

plt.plot(range(1,8), wcss, color='red')
plt.title('Graph of k values and WCSS')
plt.xlabel('k values')
plt.ylabel('wcss values')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil=[]
cl=[2,3,4,5,6]
for k in cl:
    mod=KMeans(k)
    mod.fit(data_sc1)
    score=silhouette_score(data_sc1,mod.labels_)
    print(score)
    sil.append(score)
```

```python
plt.figure(figsize=(18,10))
from scipy.cluster.hierarchy import linkage , dendrogram, fcluster,cophenet
mergings = linkage(data_sc1, method='ward',metric='euclidean')
dendrogram(mergings,truncate_mode='lastp',p=100)
plt.show()
```

```python
for i in [10,15,20,21,22,25,26,27,30,35,40,100,150,190,200,250,300,350,400,425,500]:
    clusters = fcluster(mergings, i, criterion='distance')
    print('The number of cluster for the distance of', i,' is ', len(np.unique(clusters)))
```

```python
lst=[]
for i in range(1,30):
    lst.append(('PC'+str(i)))
lst
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = data_sc1.shape[1])
pca_data = pca.fit_transform(data_sc1)
pca_data=pd.DataFrame(pca_data,columns=lst)

exp_var_ratio= pca.explained_variance_ratio_
exp_var_ratio.round(3)

cum_var=exp_var_ratio[0]
itr=2 # defined as two as first pc1 variance defined outside the loop
for j in exp_var_ratio[1:]:
    cum_var=cum_var+j
    if cum_var >= 0.95:
        break
    itr=itr+1

print('The number of principle components capturing 95 percent varaition is data is : ',itr,' Varaince explained is ', cum_var)
```
