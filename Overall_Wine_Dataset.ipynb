```python
data=data.rename({0:'Alcohol',1:'Malic acid',2:'Ash',3:'Alcalinity of ash',4:'Magnesium',5:'Total phenols',6:'Flavanoids',
             7:'Nonflavanoid phenols',8:'Proanthocyanins',9:'Color intensity',10:'Hue',11:'OD280/OD315 of diluted wines',
             12:'Proline'},axis=1)
data
```

```python
data=data.rename({13:'Missing Attribute Name'},axis=1)
data
```

```python
data2=data.drop('Alcohol',axis=1)
```

```python
data2
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_sc=sc.fit_transform(data2)
```

```python
data_sc=pd.DataFrame(data2,columns=data2.columns,index=data2.index)
data_sc.head()
```

```python
clust=KMeans(n_clusters=3,random_state=1)
clust_mod=clust.fit(data_sc)
clust_mod
```

```python
labels=clust_mod.labels_
labels
```

```python
data2['Labels']=clust_mod.labels_
data2
```

```python
wcsse=[]
cl=[1,2,3,4,5,6,7,8]
for i in cl:
    mod=KMeans(n_clusters=i,random_state=0)
    mod.fit(data_sc)
    print('K:',i)
    print(mod.inertia_) #inertia value
    wcsse.append(mod.inertia_)
```

```python
plt.plot(cl,wcsse)
plt.show()
```

```python
clust=KMeans(n_clusters=4,random_state=1)
clust_mod=clust.fit(data_sc)
labels=clust_mod.labels_
labels
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_sc=sc.fit_transform(data3)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data_sc,clust_mod.labels_)
```

```python
data_sc
```

```python
sil1=[]
cl=[3,4,5,6,7,8] # 1 should not be there as we cant calculate the silhoutte score with one cluster, we need min of 2
for k in cl:       # with one cluster we can only find wcsse
    mod=KMeans(n_clusters=k)
    mod.fit(data_sc)
    sil=silhouette_score(data_sc,mod.labels_)
    sil1.append(sil)
sil1
```

```python
inp_finalData=final_data.drop('Labels',axis=1)
out_finalData=final_data['Labels']
```

```python
pca_mod=PCA(inp_finalData.shape[1]) 
pcadata=pca_mod.fit_transform(inp_finalData)
pcadata
```

```python
pcadata=pd.DataFrame(pcadata,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
pcadata
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(inp_finalData,out_finalData,test_size=0.3,random_state=48)
mod=RandomForestClassifier(random_state=0)
mod.fit(xtrain,ytrain)
ypred=mod.predict(xtest)
print('Without PCA',accuracy_score(ytest,ypred))
```
